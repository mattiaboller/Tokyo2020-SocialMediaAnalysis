```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import time
```

```python
# Creating list to append tweet data to
tweets_list = []
```

```python
since = '2021-08-07'
until = '2021-08-08'
```

```python
t0 = time.time()

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#tokyo2020 since:{0} until:{1} lang:en'.format(since, until)).get_items()):
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.user.verified, tweet.user.id, tweet.user.followersCount,
                        tweet.replyCount, tweet.retweetCount, tweet.likeCount, tweet.quoteCount,
                        tweet.lang, tweet.retweetedTweet, tweet.quotedTweet, tweet.mentionedUsers])
    
    
# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'TweetId', 'Text', 'Username', 'UserVerified','UserId', 'UserFollowers', 'ReplyCount', 
                                               'RetweetCount', 'LikeCount', 'QuoteCount',
                                                 'Language', 'Retweeted', 'Quoted', 'Mentioned'])

tweets_df.Quoted = tweets_df.Quoted.apply(lambda x: x.user.username if x is not None else None)

tweets_df.Retweeted = tweets_df.Retweeted.apply(lambda x: x.user.username if x is not None else None)

tweets_df.Mentioned = tweets_df.Mentioned.apply(lambda x: [y.username for y in x] if x is not None else None)

tweets_df.to_json('{0}.json'.format(since))

print('Scrape time: %0.3f seconds.' % (time.time() - t0))
```
